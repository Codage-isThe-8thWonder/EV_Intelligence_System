```python
import pandas as pd

from sklearn.linear_model import LinearRegression

from sklearn.ensemble import (
    RandomForestRegressor,
    ExtraTreesRegressor,
    GradientBoostingRegressor
)

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
```

```python
df = pd.read_csv("../datasets/EVProject_dataset.csv")

print(df.shape)
df.head()
```

```python
df["Component_Health_Score"].value_counts()
```

```python
df["Component_Health_Score"].nunique()
```

```python
features = [
    "SoC",
    "SoH",
    "Battery_Voltage",
    "Battery_Current",
    "Battery_Temperature",
    "Charge_Cycles",
    "Motor_Temperature",
    "Motor_Vibration",
    "Motor_Torque",
    "Motor_RPM",
    "Power_Consumption",
    "Brake_Pad_Wear",
    "Brake_Pressure",
    "Reg_Brake_Efficiency",
    "Tire_Pressure",
    "Tire_Temperature",
    "Suspension_Load",
    "Ambient_Temperature",
    "Ambient_Humidity",
    "Load_Weight",
    "Driving_Speed",
    "Distance_Traveled",
    "Idle_Time",
    "Route_Roughness",
    "Maintenance_Type"
]

target = "Component_Health_Score"
```

```python
X = df[features]

y = df[target]

split_idx = int(len(df) * 0.8)

X_train = X.iloc[:split_idx]
X_test = X.iloc[split_idx:]

y_train = y.iloc[:split_idx]
y_test = y.iloc[split_idx:]
```

```python
models = {

    "Linear Regression":
        LinearRegression(),

    "Random Forest":
        RandomForestRegressor(
            n_estimators=200,
            random_state=42,
            n_jobs=-1
        ),

    "Extra Trees":
        ExtraTreesRegressor(
            n_estimators=200,
            random_state=42,
            n_jobs=-1
        ),

    "Gradient Boosting":
        GradientBoostingRegressor(
            n_estimators=200,
            random_state=42
        )
}
```

```python
results = []

for name, model in models.items():

    print(f"Training {name}...")

    model.fit(
        X_train,
        y_train
    )

    y_pred = model.predict(
        X_test
    )

    mae = mean_absolute_error(
        y_test,
        y_pred
    )

    mse = mean_squared_error(
        y_test,
        y_pred
    )

    rmse = mse ** 0.5

    r2 = r2_score(
        y_test,
        y_pred
    )

    results.append([
        name,
        mae,
        rmse,
        r2
    ])
```

```python
results_df = pd.DataFrame(
    results,
    columns=[
        "Model",
        "MAE",
        "RMSE",
        "R2"
    ]
)

results_df.sort_values(
    by="R2",
    ascending=False
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression

pipeline = Pipeline([
    (
        "imputer",
        SimpleImputer(strategy="median")
    ),
    (
        "model",
        LinearRegression()
    )
])

pipeline.fit(X_train, y_train)
```

```python
import joblib

joblib.dump(
    pipeline,
    "../saved_models/component_health_model.pkl"
)

print("Component Health Model Saved Successfully")
```
