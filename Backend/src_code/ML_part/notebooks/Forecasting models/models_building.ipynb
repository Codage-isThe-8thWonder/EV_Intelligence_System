```python
import pandas as pd

df = pd.read_csv(
    "../../datasets/EVProject_dataset.csv"
)
```

```python
df.head()
```

```python
df["Timestamp"] = pd.to_datetime(df["Timestamp"])
df = df.sort_values(by="Timestamp").reset_index(drop=True)
```

```python
df["soc_1h"] = df["SoC"].shift(-4)

df["soc_6h"] = df["SoC"].shift(-24)

df["soc_24h"] = df["SoC"].shift(-96)
```

```python
df["voltage_1h"] = df["Battery_Voltage"].shift(-4)

df["voltage_6h"] = df["Battery_Voltage"].shift(-24)

df["voltage_24h"] = df["Battery_Voltage"].shift(-96)
```

```python
ROWS_PER_DAY = 96

df["soh_30d"] = (
    df["SoH"].shift(
        -(30 * ROWS_PER_DAY)
    )
)

df["soh_90d"] = (
    df["SoH"].shift(
        -(90 * ROWS_PER_DAY)
    )
)

df["soh_180d"] = (
    df["SoH"].shift(
        -(180 * ROWS_PER_DAY)
    )
)
```

```python
FEATURE_COLUMNS = [
    "SoC",
    "SoH",
    "Battery_Voltage",
    "Battery_Current",
    "Power_Consumption",
    "Driving_Speed"
]
```

### Building model for SoC

```python
SOC_TARGETS = [
    "soc_1h",
    "soc_6h",
    "soc_24h"
]
```

```python
soc_df = df.dropna(subset=SOC_TARGETS).copy()
```

```python
print(soc_df.shape)
```

```python
split_idx = int(
    len(soc_df) * 0.8
)

train_df = soc_df.iloc[:split_idx]

test_df = soc_df.iloc[split_idx:]
```

```python
print(train_df.shape)

print(test_df.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
import joblib

feature_scaler = MinMaxScaler()

train_df[FEATURE_COLUMNS] = (
    feature_scaler.fit_transform(
        train_df[FEATURE_COLUMNS]
    )
)

test_df[FEATURE_COLUMNS] = (
    feature_scaler.transform(
        test_df[FEATURE_COLUMNS]
    )
)

joblib.dump(
    feature_scaler,
    "soc_feature_scaler.pkl"
)
```

```python
target_scaler = MinMaxScaler()

train_df[SOC_TARGETS] = (
    target_scaler.fit_transform(
        train_df[SOC_TARGETS]
    )
)

test_df[SOC_TARGETS] = (
    target_scaler.transform(
        test_df[SOC_TARGETS]
    )
)

joblib.dump(
    target_scaler,
    "soc_target_scaler.pkl"
)
```

```python
X_train = train_df[
    FEATURE_COLUMNS
].values.astype("float32")

y_train = train_df[
    SOC_TARGETS
].values.astype("float32")

X_test = test_df[
    FEATURE_COLUMNS
].values.astype("float32")

y_test = test_df[
    SOC_TARGETS
].values.astype("float32")
```

```python
import numpy as np

WINDOW_SIZE = 24

def create_sequences(
    X,
    y,
    window_size
):

    X_seq = []
    y_seq = []

    for i in range(
        window_size,
        len(X)
    ):

        X_seq.append(
            X[
                i-window_size:i
            ]
        )

        y_seq.append(
            y[i]
        )

    return (
        np.array(X_seq),
        np.array(y_seq)
    )
```

```python
X_train_seq, y_train_seq = (
    create_sequences(
        X_train,
        y_train,
        WINDOW_SIZE
    )
)

X_test_seq, y_test_seq = (
    create_sequences(
        X_test,
        y_test,
        WINDOW_SIZE
    )
)
```

```python
X_train_seq.shape, y_train_seq.shape, X_test_seq.shape, y_test_seq.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    LSTM,
    Dense
)

model = Sequential([
    LSTM(
        32,
        input_shape=(
            X_train_seq.shape[1],
            X_train_seq.shape[2]
        )
    ),

    Dense(
        16,
        activation="relu"
    ),

    Dense(3)
])
```

```python
model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)
```

```python
history = model.fit(
    X_train_seq,
    y_train_seq,
    epochs=5,
    batch_size=32,
    validation_data=(
        X_test_seq,
        y_test_seq
    )
)
```

```python
predictions = model.predict(
    X_test_seq[:10]
)
```

```python
model.save(
    "soc_forecast_model.keras"
)
```

```python
actual_predictions = (
    target_scaler.inverse_transform(
        predictions
    )
)

actual_targets = (
    target_scaler.inverse_transform(
        y_test_seq[:10]
    )
)
```

```python
import pandas as pd

comparison_df = pd.DataFrame({
    "Actual_1H": actual_targets[:, 0],
    "Predicted_1H": actual_predictions[:, 0],

    "Actual_6H": actual_targets[:, 1],
    "Predicted_6H": actual_predictions[:, 1],

    "Actual_24H": actual_targets[:, 2],
    "Predicted_24H": actual_predictions[:, 2],
})

comparison_df
```

```python
preds = model.predict(
    X_test_seq[:1000]
)

preds = target_scaler.inverse_transform(preds)

actual = target_scaler.inverse_transform(
    y_test_seq[:1000]
)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

for i, horizon in enumerate(
    ["1H", "6H", "24H"]
):

    rmse = np.sqrt(
        mean_squared_error(
            actual[:, i],
            preds[:, i]
        )
    )

    print(
        horizon,
        "RMSE =",
        rmse
    )
```
