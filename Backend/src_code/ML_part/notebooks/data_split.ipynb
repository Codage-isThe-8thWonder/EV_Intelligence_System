```python
import pandas as pd
df = pd.read_csv("../datasets/EV_Predictive_Maintenance_Dataset_15min.csv")
```

```python
df["Timestamp"] = pd.to_datetime(df["Timestamp"])

df = df.sort_values("Timestamp")

df.reset_index(drop=True, inplace=True)
```

```python
df.info()
```

```python
split_idx = int(len(df) * 0.85)

train_df = df.iloc[:split_idx]

simulation_df = df.iloc[split_idx:]
```

```python
print(train_df.shape)
print(simulation_df.shape)
```

```python
train_df.to_csv(
    "../datasets/EVProject_dataset.csv",
    index=False
)

simulation_df.to_csv(
    "../datasets/simulation_dataset.csv",
    index=False
)
```
